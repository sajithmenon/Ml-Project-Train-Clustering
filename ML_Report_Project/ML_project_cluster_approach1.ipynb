COABLCKSCTN column helps us to know the sequence of next station visited by train, so we will use this to our advantage and try to cluster based on this. But since we can't use this feature directly, we seperate it into unique stations and then assign values (as 1) to those stations which are being visited and (0) to those stations which are not being visited.

Example :- We have 4 stations : A,B,C,D and the train visits A->B->C and doesn't visit D.
So the method we used is

Train number    A  B  C  D
xyz             1  1  0  0
xyz             0  1  1  0

Doing so helps us to get an idea of sequence of stations visited by trains.

```
# Using silhoutte score to find optimal number of cluster
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
sil = []
kmax = 30

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(test_df)
  labels = kmeans.labels_
  sil.append(silhouette_score(test_df, labels, metric = 'euclidean'))
```

```
plt.plot(range(29),sil)
plt.show
```

```
# Using elbow method to find optimal number of cluster
kmeans_kwargs = {"init": "random", "n_init": 10, "max_iter": 300,"random_state": 42}
sse = []
for k in range(2, 25):
  kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
  kmeans.fit(test_df)
  sse.append(kmeans.inertia_)
```

```
plt.plot(range(2, 25), sse)
plt.xticks(range(2, 25))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

Optimal clusters must be around 20
